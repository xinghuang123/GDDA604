```python
# Task A, a, Load the dataset into a DataFrame.

# Import the necessary library
import pandas as pd

# Define the file path
file_path = 'Sales_Sample_Public_Dataset.csv'

# Load the dataset into a DataFrame
sales_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
sales_data.head()
```

```python
# Task A, c, Apply three (3) operations to handle missing values in a dataset. (6 Marks)

# 1. Dropping rows with missing values  

# Step 1: Check for missing values
print("\nMissing values before handling: ")
print(sales_data.isnull().sum())
```

```python
# Step 2: Determine the threshold

# Show the number of column in the DataFrame
num_columns = sales_data.shape[1]
print(f"Number of columns in the DataFrame: {num_columns}")
```

```python
# There are 25 columns in total, 
# we will only keep the rows with number of non-NaN value equals or is greater than 23  
# Set the threshold = 23 
threshold_value = 23 # Set the threshold = 23
```

```python
# Step 3: Apply the dropna() method
cleaned_data_drop = sales_data.dropna(thresh = threshold_value)
print("\nMissing values after dropping: ")
print(cleaned_data_drop.isnull().sum())
```

```python
# Task A, c, Apply three (3) operations to handle missing values in a dataset. (6 Marks)

# 2. Filling missing values with a specific value

import numpy as np

# Check for missing values in the POSTALCODE column
print("Missing values in 'POSTALCODE' before filling:")
print(sales_data['POSTALCODE'].isnull().sum())

# Define the postal codes for the cities
san_francisco_postal_code = '94102'
los_angeles_postal_code = '90001'

# Use numpy.where to fill missing postal codes based on the city
sales_data['POSTALCODE'] = np.where(
    (sales_data['CITY'] == 'San Francisco') & (sales_data['POSTALCODE'].isnull()), 
    san_francisco_postal_code, 
    sales_data['POSTALCODE']
)

sales_data['POSTALCODE'] = np.where(
    (sales_data['CITY'] == 'Los Angeles') & (sales_data['POSTALCODE'].isnull()), 
    los_angeles_postal_code, 
    sales_data['POSTALCODE']
)

# Check for missing values in the POSTALCODE column after filling
print("Missing values in 'POSTALCODE' after filling:")
print(sales_data['POSTALCODE'].isnull().sum())

# Display the first few rows to verify the changes
print(sales_data.head())
```

```python
# Task A, c, Apply three (3) operations to handle missing values in a dataset. (6 Marks)

# 3. Imputing missing values with forward/backward fill

import numpy as np

# Check for missing values before filling
missing_values_before = sales_data.isnull().sum()
print("Missing values before forward/backward fill:")
print(missing_values_before[missing_values_before > 0])

# Apply forward fill
sales_data_ffill = sales_data.ffill()

# Check for missing values after forward fill
missing_values_after_ffill = sales_data_ffill.isnull().sum()
print("\nMissing values after forward fill:")
print(missing_values_after_ffill[missing_values_after_ffill > 0])

# Apply backward fill
sales_data_bfill = sales_data.bfill()

# Check for missing values after backward fill
missing_values_after_bfill = sales_data_bfill.isnull().sum()
print("\nMissing values after backward fill:")
print(missing_values_after_bfill[missing_values_after_bfill > 0])

# Optionally, we can combine forward fill and backward fill
sales_data_combined_fill = sales_data.ffill().bfill()

# Check for missing values after combined fill
missing_values_after_combined = sales_data_combined_fill.isnull().sum()
print("\nMissing values after combined fill:")
print(missing_values_after_combined[missing_values_after_combined > 0])

# Display the first few rows to verify the changes
sales_data_combined_fill.head()
```

```python
# Task A, d, Choose a column and perform the sorting technique. (6 Marks)

# Sort the dataset by the SALES column in ascending order
sorted_sales_asc = sales_data_combined_fill.sort_values(by='SALES', ascending=True)

# Display the first few rows of the sorted dataset (ascending)
print("Dataset sorted by SALES (ascending):")
print(sorted_sales_asc.head())

# Sort the dataset by the SALES column in descending order
sorted_sales_desc = sales_data_combined_fill.sort_values(by='SALES', ascending=False)

# Display the first few rows of the sorted dataset (descending)
print("\nDataset sorted by SALES (descending):")
print(sorted_sales_desc.head())
```

```python
# Task A, e, Define a condition to filter transactions from the dataset. (6 Marks)

# Define the condition to filter transactions
condition = (sales_data_combined_fill['SALES'] > 5000) & (sales_data_combined_fill['STATUS'] == 'Shipped')

# Apply the condition to filter the dataset
filtered_transactions = sales_data_combined_fill[condition]

# Display the first few rows of the filtered dataset
print("Filtered transactions (SALES > 5000 and STATUS == 'Shipped'):")
print(filtered_transactions.head())

# Display the number of filtered transactions
print("\nNumber of filtered transactions:")
print(filtered_transactions.shape[0])
```

```python
# Task A, f, Create a new column to derive additional information. (7 Marks)

# Multiply QUANTITYORDERED and PRICEEACH to create the SALES_NEW column
sales_data_combined_fill['SALES_NEW'] = sales_data_combined_fill['QUANTITYORDERED'] * sales_data_combined_fill['PRICEEACH']

# Display the first few rows to verify the new column
print("Dataset with new 'SALES_NEW' column:")
print(sales_data_combined_fill[['QUANTITYORDERED', 'PRICEEACH', 'SALES_NEW']].head())
```

```python
# Task B, a, Group the dataset based on a categorical variable and calculate summary statistics.

import pandas as pd

# Group the dataset by PRODUCTLINE and calculate summary statistics
grouped_data = sales_data_combined_fill.groupby('PRODUCTLINE').agg(
    total_sales=('SALES', 'sum'),
    average_sales=('SALES', 'mean'),
    total_quantity_ordered=('QUANTITYORDERED', 'sum'),
    average_quantity_ordered=('QUANTITYORDERED', 'mean'),
    average_price_each=('PRICEEACH', 'mean')
).reset_index()

# Display the grouped data with summary statistics
print("Grouped data by PRODUCTLINE with summary statistics:")
print(grouped_data)
```

```python
# # Task B, e, Apply inferential statistical method to quantify the relationships between variables.

# Step 1: Calculate the correlation coefficient between “PRICEEACH” and “SALES_NEW”

import scipy.stats as stats

# Calculate correlation coefficient
correlation, p_value = stats.pearsonr(sales_data_combined_fill['PRICEEACH'], sales_data_combined_fill['SALES_NEW'])
print(f"Correlation coefficient between PRICEEACH and SALES_NEW: {correlation}")
print(f"P-value: {p_value}")
```

```python
# Step 2: Apply Ordinary Least Squares (OLS) regression model

import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy import stats
import statsmodels.api as sm

# Define the independent variable (X) and dependent variable (y)
X = sales_data_combined_fill['PRICEEACH']
y = sales_data_combined_fill['SALES_NEW']

# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())
```

```python
# Hypothesis Testing

# Extract the p-value for the PRICEEACH coefficient from the model summary
p_value_priceeach = model.pvalues['PRICEEACH']
print(f"P-value for PRICEEACH: {p_value_priceeach}")

# Determine significance level (alpha)
alpha = 0.05

# Hypothesis testing
if p_value_priceeach < alpha:
    print("Reject the null hypothesis: PRICEEACH is a significant predictor of SALES_NEW.")
else:
    print("Fail to reject the null hypothesis: PRICEEACH is not a significant predictor of SALES_NEW.")
```
